```python
##abs() - 절대값 계산 함수 , round() - 반올림 계산 함수 , trunc() - 버림 계산 함수 

a = math.ceil(324.6) ## 소수점 올림
b = round(1.567,2) ## 셋째 자리에서 반올림
c = round(1.5) #소수점 반올림
d = math.trunc(1.7)  # 내림
e = abs(10) # 절대 값
f = abs(-10) # 절대 값


print('a:',a) 
print('b:',b)
print('c:',c)
print('d:',d)
print('e:',e)
print('f:',f)
```

```python
##os.path : 문자열
## pathlib 객체
## 연산자를 새롭게 정의할 수 있게됨

import os
from pathlib import Path

dir_name = 'dir'
sub_dir_name = 'sub_dir_name'
file_name = 'file'

# 기존 (os.path)
file = os.path.join(dir_name, sub_dir_name, file_name)
print(file)


# pathlib  ## 보기 편하다
dir = Path(dir_name)
file = dir / sub_dir_name / file_name
print(file)
```

```python
## 파일 쓰기/읽기


from pathlib import Path

filename = 'Untitled1-0611_17'


# pathlib 사용 전   
f = open(filename, 'w+')   ## 쓰기
f.write('파일에 기록될 문자열입니다.')
f.close()


file = open(filename, 'r')   ## 읽기
r = file.read()
file.close()
print(r)






# pathlib 사용 후
path = Path(filename)
r = path.write_text('파일쓰기1')   ## 파일 쓰기
r = path.read_text()     ## 파일 읽기
print(r)

```

```python
## fileinput 
## 파일 내용 변경

import fileinput
from pathlib import Path


filename = "fileinput.txt"


with fileinput.FileInput(filename, inplace = True, backup = '.bak') as f:
    
    for line in f:
        print(line)
        if "ryan" in line:
            print(line.replace("라이언", "ryan"), end = '') ## ryan을 '라이언'으로 변경
            ##print("라이언", end ='\n')   ## ryan이 포함된 줄을 '라이언'으로 변경
        
        else:
            print(line, end = '')


            
            
            
path = Path(filename)
r = path.read_text()     ## 파일 읽기
print(r)
            
            
```

```python
## glob
## 특정 파일만 출력 

import glob
for filename in glob.glob('glob.*'):
    print(filename)
```

```python
## iglob
## 원하는 폴더 안의 특정 파일 부르기
## (recursive) * - 모든 서브 디렉토리를 포함하다

for filename in glob.iglob('**/*.txt', recursive=True):
    print(filename)

```

```python
## pickle
## 1) 자료형의 변경 없이 파일로 저장하여 로드 가능 
## 2) 데이터를 저장하거나 불러올 때는 파일을 Byte형식으로 읽거나 씀
## 3) 모든 파이써 데이터 객체를 저장하고 읽기 가능


import pickle

list = ['a','b','c']
        
## dump pickle
with open("data.pickle","wb") as fw:
        pickle.dump(list,fw)
        
## Load pickle
with open("data.pickle","rb") as fr:
        data = pickle.load(fr)
        
print(data)
```
